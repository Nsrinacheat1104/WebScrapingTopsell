```python
import requests
from bs4 import BeautifulSoup

# URL dari halaman Laptop & Notebook di Top Sell Belanja
url = 'https://www.topsellbelanja.com/product-category/laptop-notebook/'

# Mengirim permintaan ke halaman
response = requests.get(url)

# Memastikan permintaan berhasil (status code 200)
if response.status_code == 200:
    # Parsing konten HTML dari halaman
    bs = BeautifulSoup(response.content, 'html.parser')

    # Menemukan semua elemen yang berisi produk
    products = bs.find_all('li', class_='product')

    # Iterasi melalui produk yang ditemukan
    for product in products:
        # Mendapatkan nama produk
        title = product.find('h2', class_='woocommerce-loop-product__title').text.strip()

        # Mendapatkan harga produk
        price = product.find('span', class_='woocommerce-Price-amount').text.strip()

        # Mendapatkan URL produk
        link = product.find('a')['href']

        # Menampilkan hasil
        print(f"Product: {title}\nPrice: {price}\nLink: {link}\n")
else:
    print(f"Error: Unable to fetch the page, status code {response.status_code}")
```
